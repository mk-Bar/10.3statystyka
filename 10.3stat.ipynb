Stwórz dane z rozkładu normalnego o średniej 3 i odchyleniu standardowym 1 dla 100 oraz 10 000. Na podstawie tych sztucznie stworzonych zbiorów oblicz statystyki poznane w module statystki opisowej. Opisz różnice pomiędzy statystkami i wynik swojej pracy wyślij Mentorowi jako Jupyter Noteboook zapisany w Github

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
data_100 = np.random.normal(loc=3, scale=1, size=100)
data_10000 = np.random.normal(loc=3, scale=1, size=10000)

data_100=pd.Series(data_100)

# Funkcja obliczająca statystyki opisowe
def descriptive_stats(data):
    dataPd=pd.Series(data)
    return {
        "Liczba elementów": len(data),
        "Średnia": dataPd.mean(),
        "Mediana": dataPd.median(),
        "ModaPd": dataPd.mode(),
        "Moda": stats.mode(dataPd),
        "Minimum": dataPd.min(),
        "Maksimum": dataPd.max(),
        "1. kwartyl": dataPd.quantile(0.25),
        "3. kwartyl": dataPd.quantile(0.75),
        "Zakres": dataPd.max()-dataPd.min(),
        "Rozstęp międzykwartylowy (IQR)": dataPd.quantile(0.75)- dataPd.quantile(0.25),
        "Wariancja": dataPd.var(),
        "SkośnośćPd": dataPd.skew()
    }

# Obliczanie statystyk
stats_100 = descriptive_stats(data_100)
stats_10000 = descriptive_stats(data_10000)

# Tworzenie DataFrame dla porównania
df_stats = pd.DataFrame([stats_100, stats_10000], index=["N=100", "N=10000"])

```
